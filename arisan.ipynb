```python
import random
import matplotlib.pyplot as plt

# Jumlah peserta arisan
N = 20  # Misalnya 20 peserta

# Inisialisasi daftar peserta
peserta = list(range(1, N + 1))
pemenang = []

# Simulasi pengocokan
while peserta:
    nomor_terpilih = random.choice(peserta)
    pemenang.append(nomor_terpilih)
    peserta.remove(nomor_terpilih)

# Tampilan hasil
print("Urutan Pemenang Arisan:", pemenang)

# Visualisasi
plt.figure(figsize=(10, 6))
plt.bar(range(1, N + 1), pemenang, color='skyblue')
plt.xticks(range(1, N + 1))
plt.xlabel("Urutan Pengundian")
plt.ylabel("Nomor Peserta")
plt.title("Simulasi Sistem Koclok Arisan")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
